```python
with open("sensorData.txt", mode = 'r') as file:
    for line in file:
        if line.strip() == 'None':
            continue # skip if none
        try:
            line = float(line.strip())
        except ValueError:
            print(f"Skipping non-numeric line: {line}")
            continue
        # print(f"checking for value {line}")
        if line > 100.00:
            print(f"Danger: {line}")
        elif 50.00 < line <= 100.00:
            print(f"Warning: {line}")
        elif line == 0.0:
            break
    else:
        print("All readings processed safely")

```

```python
def decorator_fun(fun):
    def wrapperFun(*args):
        print("Brfore calling the function")
        result = fun(*args)
        print("After calling the function")
        return result
    return wrapperFun
        
```

```python
@decorator_fun
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
```

```python
import datetime
def log_execution(fun):
    def wrapperFun(*args, **kwargs):
        # log  start time
        print(f"[LOG] calling {fun.__name__}")
        print(f"[LOG] Arguments: Positional={args}, Named={kwargs}")
        print(f"Start time: {datetime.datetime.now()}")
        result = fun(*args, **kwargs)
        print(f"End time: {datetime.datetime.now()}")
        return result
    return wrapperFun
@log_execution
def compute_sum(a, b):
    return a + b
    
# Call the function outside of its own definition
print(f"Final Result: {compute_sum(5, 7)}")
```

```python
class BankAccount:

    #initializer method as constructor
    def __init__(self, account_number,balance=0.0):
        self.account_number = account_number
        self.__balance = balance # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")
    
    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient funds.")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance: {self.__balance}")
        
    def get_balance(self):
        if self.__balance > 0:
            print(f"Current balance: {self.__balance}")
        else:
            print("Your account balance is zero or negative.")


cust1 = BankAccount("1234567890", 1000.0)
cust1.get_balance()
```

Calculate Revenue with decorator function
    

```python
import datetime
import csv
def process_logger(func):
    def wrapper_fun(*args, **kwargs):
        print(f"---Starting Processing for {func.__name__}---")
        print(f"------Start time------")
        print(datetime.datetime.now())
        result = func(*args, **kwargs)
        print(f"------End time------")
        print(datetime.datetime.now()) 
        print(f"---Finished Processing for {func.__name__}---")
        return result
    return wrapper_fun

@process_logger
def calculate_revenue(file_path):
    total_revenue = 0.0
    with open(file_path, mode ='r') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = next(csv_reader)  # Skip header
        for row in csv_reader:
            tid, product, price, quantity = row
            # skip if quantity is missing
            if quantity == 'None':
                print(f"Skipping transaction {tid} due to missing quantity.")
                continue
            # type cast the string into number
            price = float(price)
            quantity = int(quantity)
            
            if quantity == 0:
                print(f"Skipping transaction {tid} due to zero quantity.")
                continue

            total_revenue += price * quantity
            print(f"Processed {product}: ${price * quantity}")
    return total_revenue


# Execute the process
result = calculate_revenue(file_path='./salesData.csv')
print(f"Total Revenue Calculated: ${result}")
```
