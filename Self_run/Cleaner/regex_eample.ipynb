value = "Hello, World! This is a test. #Python @2024"

# Remove special characters (keep letters and spaces)
re.sub(r'[^a-zA-Z ]', '', value)


# r'[^a-zA-Z ] - This is the regex pattern. It says: "Match any character that is NOT a letter (a-z or A-Z) or a space."

# r: This stands for "raw string." It tells Python, "Don't try to interpret backslashes, just give this text straight to the Regex engine."
# ^ (The Hat): Inside brackets, this symbol means NOT. So [^a-zA-Z ] means "match any character that is NOT a letter or space."

# (The Space): This is the most important part! By putting a literal space in the brackets, we tell the regex NOT to delete spaces.

# The Translation: "Find every character that is NOT a letter and NOT a space, and delete it."

# Regex Pattern	    What it keeps	            What it deletes
# r'[a-zA-Z]'	    Nothing (it finds letters)	If used in sub, it deletes your text!
# r'[^a-zA-Z]'	    Letters	                    Deletes spaces, numbers, and symbols.
# r'[^a-zA-Z ]'	    Letters and Spaces	        Deletes numbers and symbols.
# r'[^a-zA-Z0-9 ]'	Letters, Numbers, Spaces	Deletes only special symbols.
