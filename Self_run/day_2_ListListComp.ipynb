```python
api_responses = [
    {"id": 1, "choices": [{"text": "Hello!", "score": 0.9}]},
    {"id": 2, "choices": [{"text": "Python is cool", "score": 0.85}]},
    {"id": 3, "choices": [{"text": "Error 404", "score": 0.1}]}
]
textls = []
for item in api_responses:
    inner = item.get('choices')
    for item in inner:
        textls.append(item.get('text'))
print((textls))

values = [item_1.get('text') for item in api_responses for item_1 in item.get('choices')]
print(values)
```

```python
# Intermediate Challenge 7: The "Word Counter" (Dictionary + Loop)
words = ["python", "ai", "python", "ml", "ai", "python"]

word_counts = {}

for item in words:
    counter = 0
    if not item in word_counts.keys():
        word_counts[item] = 1
    elif item in word_counts.keys():
        counter = word_counts.get(item) + 1
        word_counts[item] = counter

print(word_counts)
```
