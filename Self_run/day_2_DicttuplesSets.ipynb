```python
# Challenge 4: The Frequency Counter (The "Boss" Level)
text = "python is fun and python is fast"
word_count= {}
splitted = text.split(" ")
counter = 0
for value in splitted:
    if value in word_count.keys():
        word_count[value] += counter +1
    
    elif not value in word_count.keys():
        word_count[value] = 1
        
print(word_count)





```
