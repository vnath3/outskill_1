### Gemini Usage for LLM ###
**Library Need**  
To use the APIs provided by the Geimi platform, the library 'Google GenAI' is required.  
Ensure to have performed 'pip install google-genai dotenv'  
The API key from your Google AI account would be required to execute this code


```python
import os
from google import genai
from dotenv import load_dotenv

# 1. Load the .env file
load_dotenv()

# 2. Get the key (Check if you spelled it GEMINI or GEMENI in your .env file!)
api_key_value = os.environ.get("GEMINI_API_KEY")

# 3. Initialize the client correctly using the api_key= keyword
client = genai.Client(api_key=api_key_value)

# Quick check
print("Client initialized!")
```

**Invoke for chat Completion**  
>We try Gemini to get response from LLM  
>Note the differece in terms of API  

```python
response = client.models.generate_content(
                model="gemini-2.5-flash", 
                contents="I am not a hardware specialist. Explain GPU vs TPU to me."
            )
print(response.text)
```
